```python
import numpy as np
import pandas as pd
```

```python
df_imdb = pd.read_csv("simpsons_imdb.csv", sep=",", index_col=0)
df_imdb["airdates"] = df_imdb["airdates"].replace("22 Dec. 1990", "17 Dec. 1989")
df_imdb
```

```python
df_imdb["date_imdb"] = df_imdb.airdates.str.replace(' Jan. ', '/01/', regex=True).replace(' Feb. ', '/02/', regex=True).replace(' Mar. ', '/03/', regex=True).replace(' Apr. ', '/04/', regex=True).replace(' May ', '/05/', regex=True).replace(' Jun. ', '/06/', regex=True).replace(' Jul. ', '/07/', regex=True).replace(' Aug. ', '/08/', regex=True).replace(' Sep. ', '/09/', regex=True).replace(' Oct. ', '/10/', regex=True).replace(' Nov. ', '/11/', regex=True).replace(' Dec. ', '/12/', regex=True)
df_imdb.date_imdb = pd.to_datetime(df_imdb.date_imdb,format='%d/%m/%Y',errors='coerce')
df_imdb
```

```python
df_episodes = pd.read_csv("simpsons_episodes.csv", sep=",")
df_episodes = df_episodes.drop(columns=["image_url", "production_code", "video_url"])
df_episodes['date_kaggle'] = pd.to_datetime(df_episodes.original_air_date,infer_datetime_format=True,errors='coerce')
df_episodes
```
