```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links_real_world_characters = ["https://simpsons.fandom.com/wiki/Category:Real_world_characters", "https://simpsons.fandom.com/wiki/Category:Real_world_characters?from=Isabel+Sanford+%28character%29", "https://simpsons.fandom.com/wiki/Category:Real_world_characters?from=Tom+Petty+%28character%29"]
```

```python
real_world_characters_arr = []
for link in links_real_world_characters:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    real_world_characters = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in real_world_characters:
        real_world_characters_text = i.text
        real_world_characters_arr += [real_world_characters_text]
print(len(real_world_characters_arr))
real_world_characters_arr
```

```python
links_self_voicing_guest_stars = ["https://simpsons.fandom.com/wiki/Category:Self-voicing_guest_stars", "https://simpsons.fandom.com/wiki/Category:Self-voicing_guest_stars?from=Mary-Kate+and+Ashley+Olsen"]
```

```python
self_voicing_guest_stars_arr = []
for link in links_self_voicing_guest_stars:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    self_voicing_guest_stars = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in self_voicing_guest_stars:
        self_voicing_guest_stars_text = i.text
        self_voicing_guest_stars_arr += [self_voicing_guest_stars_text]
print(len(self_voicing_guest_stars_arr))
self_voicing_guest_stars_arr
```

```python
df_real_world_characters = pd.DataFrame(real_world_characters_arr, columns = ['real_world_characters'])
```

```python
df_self_voicing_guest_stars = pd.DataFrame(self_voicing_guest_stars_arr, columns = ['self_voicing_guest_stars'])
```
