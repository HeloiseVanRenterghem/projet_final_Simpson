```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links_babies = ["https://simpsons.fandom.com/wiki/Category:Babies"]
links_children = ["https://simpsons.fandom.com/wiki/Category:Children", "https://simpsons.fandom.com/wiki/Category:Children?from=Girl+from+Homer%27s+childhood", "https://simpsons.fandom.com/wiki/Category:Children?from=Playground+ghost", "https://simpsons.fandom.com/wiki/Category:Children?from=Wesley+Wiggum"]
links_teenagers = ["https://simpsons.fandom.com/wiki/Category:Teenagers"]
links_adults = ["https://simpsons.fandom.com/wiki/Category:Adults", "https://simpsons.fandom.com/wiki/Category:Adults?from=Bauer%2C+Jack%0AJack+Bauer", "https://simpsons.fandom.com/wiki/Category:Adults?from=Calliope+Juniper", "https://simpsons.fandom.com/wiki/Category:Adults?from=Cousins", "https://simpsons.fandom.com/wiki/Category:Adults?from=Eloise", "https://simpsons.fandom.com/wiki/Category:Adults?from=Gillis%2C+Stephanie%0AStephanie+Gillis", "https://simpsons.fandom.com/wiki/Category:Adults?from=Hugh+Hefner", "https://simpsons.fandom.com/wiki/Category:Adults?from=Kathy", "https://simpsons.fandom.com/wiki/Category:Adults?from=Major+Nougat", "https://simpsons.fandom.com/wiki/Category:Adults?from=Mountain+Ranger+2", "https://simpsons.fandom.com/wiki/Category:Adults?from=Parking+Lot+Guard", "https://simpsons.fandom.com/wiki/Category:Adults?from=Referee+%28Homer+Loves+Flanders%29", "https://simpsons.fandom.com/wiki/Category:Adults?from=Simpson%2C+Eliza%0AEliza+Simpson", "https://simpsons.fandom.com/wiki/Category:Adults?from=Teller+%28character%29", "https://simpsons.fandom.com/wiki/Category:Adults?from=Wickerbottom%2C+Philippa%0APhilippa+Wickerbottom"]
links_seniors = ["https://simpsons.fandom.com/wiki/Category:Seniors", "https://simpsons.fandom.com/wiki/Category:Seniors?from=Hippie+Bus+Driver", "https://simpsons.fandom.com/wiki/Category:Seniors?from=Sigurðardóttir%2C+Sigr%C3%ADn%0ASigr%C3%ADn+Siguðardóttir"]
```

```python
babies_arr = []
for link in links_babies:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    babies = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in babies:
        babies_text = i.text
        babies_arr += [babies_text]
print(len(babies_arr))
babies_arr
```

```python
children_arr = []
for link in links_children:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    children = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in children:
        children_text = i.text
        children_arr += [children_text]
print(len(children_arr))
children_arr
```

```python
teenagers_arr = []
for link in links_teenagers:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    teenagers = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in teenagers:
        teenagers_text = i.text
        teenagers_arr += [teenagers_text]
print(len(teenagers_arr))
teenagers_arr
```

```python
adults_arr = []
for link in links_adults:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    adults = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in adults:
        adults_text = i.text
        adults_arr += [adults_text]
print(len(adults_arr))
adults_arr
```

```python
seniors_arr = []
for link in links_seniors:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    seniors = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for i in seniors:
        seniors_text = i.text
        seniors_arr += [seniors_text]
print(len(seniors_arr))
seniors_arr
```

```python
df_babies = pd.DataFrame({'characters':babies_arr})
df_babies["categ"] = ["Babies"] * 74
df_children = pd.DataFrame({'characters':children_arr})
df_children["categ"] = ["Children"] * 619
df_teenagers = pd.DataFrame({'characters':teenagers_arr})
df_teenagers["categ"] = ["Teenagers"] * 101
df_adults = pd.DataFrame({'characters':adults_arr})
df_adults["categ"] = ["Adults"] * 2880
df_seniors = pd.DataFrame({'characters':seniors_arr})
df_seniors["categ"] = ["Seniors"] * 488
df = pd.concat([df_babies, df_children, df_teenagers, df_adults, df_seniors])
df = df[~df.characters.str.contains("Category:")]
df
```
