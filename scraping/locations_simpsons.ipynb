```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://simpsons.fandom.com/wiki/Category:Locations"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
all_a = soup.find_all("a", {"class": "category-page__member-link"})
all_href = [tag['href'] for tag in all_a]
all_links = [x for x in all_href if "Category:" in x]
all_links_complete = ["https://simpsons.fandom.com/" + end_url for end_url in all_links]
all_links_complete
```

```python
location_categ_arr = []
for link in all_links_complete:
    categ = requests.get(link)
    soup_categ = BeautifulSoup(categ.text, "html.parser")
    name_categ = soup_categ.find("h1", {"id": "firstHeading"}).text.replace("\n\t\t\t\t\t","").replace("\t\t\t\t","")
    locations = soup_categ.find_all("a", {"class": "category-page__member-link"})
    for location in locations:
        location_text = location.text
        location_categ = name_categ + ":" + location_text
        location_categ_arr += [location_categ]
location_categ_arr
```

```python
location_categ_series = pd.Series(location_categ_arr)
df_locations = pd.DataFrame(location_categ_series, columns=['categ_location'])
df_locations
```

```python
df_locations = df_locations[df_locations['categ_location'].str.contains('Category:')==False]
len(df_locations)
```

```python
df_locations = df_locations['categ_location'].str.split(":", n=1, expand=True)
```

```python
df_locations = df_locations.rename(columns={df_locations.columns[0]: 'categ_location', df_locations.columns[1]: 'location'})
df_locations
```
