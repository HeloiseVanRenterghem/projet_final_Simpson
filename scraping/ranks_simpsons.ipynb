```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.looper.com/757803/best-simpsons-characters-of-all-time-ranked/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
all_ranks = soup.find_all('h2')
all_ranks = [rank.text for rank in all_ranks]
all_ranks
```

```python
df_ranks = pd.DataFrame(all_ranks, columns=["all_ranks"])
df_ranks
```

```python
df_ranks["rank"] = df_ranks["all_ranks"].str[:2].str.replace(".","")
df_ranks
```

```python
df_ranks["character"] = df_ranks["all_ranks"].str[3:]
df_ranks
```

```python
df_ranks = df_ranks.drop(columns=["all_ranks"])
df_ranks
```

```python
df_ranks = df_ranks.set_index("rank")
df_ranks
```
