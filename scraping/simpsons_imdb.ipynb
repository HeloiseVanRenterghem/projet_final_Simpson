```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pages_seasons_links = ["https://www.imdb.com/title/tt0096697/episodes?season=" + str(x) for x in range(1, 34)]
pages_seasons_links
```

```python
episodes_titles = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    episode_title = [tag['title'] for tag in soup.select('strong a[title]')]
    episodes_titles += [episode_title]
episodes_titles
```

```python
episodes_summaries = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    for i in soup.select('div.info div.item_description'):
        episode_summary = i.text
        episodes_summaries += [episode_summary]
episodes_summaries
```

```python
episodes_rankings = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    for i in soup.select('div.info div.ipl-rating-widget div[class="ipl-rating-star small"] span.ipl-rating-star__rating'):
        episode_ranking = i.text
        episodes_rankings += [episode_ranking]
episodes_rankings
```

```python
episodes_total_votes = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    for i in soup.select('div.info div.ipl-rating-widget div[class="ipl-rating-star small"] span.ipl-rating-star__total-votes'):
        episode_total_votes = i.text
        episodes_total_votes += [episode_total_votes]
episodes_total_votes
```

```python
episodes_airdates = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    for i in soup.select('div.airdate'):
        episode_airdate = i.text
        episodes_airdates += [episode_airdate]
episodes_airdates
```

```python
episodes_links = []
for link in pages_seasons_links:
    page_season = requests.get(link)
    soup = BeautifulSoup(page_season.text, "html.parser")
    episode_link = [tag['href'] for tag in soup.select('strong a[href]')]
    episodes_links += [episode_link]
episodes_links
```

```python
episodes_titles = sum(episodes_titles, [])
episodes_titles
```

```python
episodes_links = sum(episodes_links, [])
episodes_links
```

```python
episodes_links = ["https://www.imdb.com" + str(x) for x in episodes_links if x.startswith('/title/')]
episodes_links
```

```python
print(len(episodes_titles))
print(len(episodes_summaries))
print(len(episodes_rankings))
print(len(episodes_total_votes))
print(len(episodes_airdates))
print(len(episodes_links))
```

```python
df_imdb = pd.DataFrame()
df_imdb["title"] = episodes_titles
df_imdb["summaries"] = episodes_summaries
df_imdb["airdates"] = episodes_airdates
df_imdb["rankings"] = episodes_rankings
df_imdb["total_votes"] = episodes_total_votes
df_imdb["links"] = episodes_links
df_imdb
```

```python
df_imdb["summaries"] = df_imdb["summaries"].str.replace("\n","")
df_imdb["summaries"] = df_imdb["summaries"].str.strip()
df_imdb["airdates"] = df_imdb["airdates"].str.replace("\n","")
df_imdb["airdates"] = df_imdb["airdates"].str.strip()
df_imdb["title"] = df_imdb["title"].str.strip()
df_imdb["total_votes"] = df_imdb["total_votes"].str.replace("(","")
df_imdb["total_votes"] = df_imdb["total_votes"].str.replace(")","")
df_imdb["total_votes"] = df_imdb["total_votes"].str.replace(",","")
df_imdb
```
