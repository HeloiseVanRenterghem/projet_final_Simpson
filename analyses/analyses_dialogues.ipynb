## Nb de mots

```python
df['word_count_int'] = pd.to_numeric(df['word_count'], errors='coerce').fillna(0).astype('int')
```

```python
print(df['word_count_int'].sum()) # nb de mots dans tte la sÃ©rie
print(df['word_count_int'].sum() / len(pd.unique(df['episode_id']))) # nb de mots moyen par Ã©p
print(df['word_count_int'].mean()) # nb de mots moyen par ligne de dialogue
```

```python
df.sort_values(by=['word_count_int'], ascending=False).head(1)
```

## Nb de mots et de lignes de dialogue par saison

```python
df_lines_grouped_ep = df.groupby(['episode_id'])['id'].unique().to_frame().rename(columns = {'id':'nb_lines'})
df_lines_grouped_ep['nb_lines'] = df_lines_grouped_ep['nb_lines'].str.len()
df_lines_grouped_ep
```

```python
df_words_grouped_ep = df.groupby(['episode_id'])['word_count_int'].sum().to_frame()
df_words_grouped_ep
```

```python
df_grouped_ep = df_lines_grouped_ep
df_grouped_ep['nb_words'] = df_words_grouped_ep['word_count_int']
df_grouped_ep['ratio'] = round(df_grouped_ep['nb_words'] / df_grouped_ep['nb_lines'])
df_grouped_ep
```

```python
df_episodes = pd.read_csv("simpsons_episodes.csv", index_col=0)
df_episodes
```

```python
df_grouped_season = df_grouped_ep.join(df_episodes)
df_grouped_season
```

```python
df_grouped_season = round(df_grouped_season.groupby('season')['nb_lines', 'nb_words'].mean())
df_grouped_season
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(df_grouped_season.index, df_grouped_season["nb_lines"], color="#F15A2A")
# set x-axis label
ax.set_xlabel("Saison")
# set y-axis label
ax.set_ylabel("Nombre moyen de lignes de dialogue par Ã©pisode", color="#F15A2A")
ax.set_ylim(150, 300)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df_grouped_season.index, df_grouped_season["nb_words"],color="#2C5B95")
ax2.set_ylabel("Nombre moyen de mots par Ã©pisode",color="#2C5B95")
ax2.set_ylim(1500, 2700)

plt.xlim([1, 26])
plt.title("Ã‰volution des dialogues", fontweight="bold")
fig.set_size_inches(12, 7)
plt.locator_params(axis='x', nbins=26)

#plt.savefig('lines_words_season.png')

plt.show()
```

## Mots les plus frÃ©quents

```python
all_words = []
for i in df["normalized_text"]:
    j = str(i).split()
    all_words += [j]
all_words
```

```python
all_words = [item for sublist in all_words for item in sublist]
all_words
```

```python
import nltk
from nltk.corpus import words, stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
import os
from os import path
```

```python
stop_words = set(stopwords)
clean_text = [word for word in all_words if word not in stop_words]
clean_text
```

```python
from collections import Counter
dict_all_words = Counter(clean_text)
dict_all_words
```

```python
d_words = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
words_mask = np.array(Image.open(path.join(d_words, '/Users/heloisevanrenterghem/Documents/THPprojets/Simpsons/all_words.png')))
words_colors = ImageColorGenerator(words_mask)
```

```python
wordcloud_words = WordCloud(font_path = 'Homer_Simpson_Revised.ttf', width=4000, height=4000, background_color="white", mask=words_mask, max_words=10000, stopwords=stopwords, max_font_size=80).generate_from_frequencies(dict_all_words)
wordcloud_words.recolor(color_func = words_colors)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud_words, interpolation='bilinear')
plt.axis("off")
plt.title("Mots les plus utilisÃ©s (sans stopwords)", fontweight="bold")
#plt.savefig('all_words.png')
plt.show()
```
